# Drop unnamed Column

```python
df=df.drop("Unnamed: 0",axis =1)
print(df.head())
```

# Change weekly study hours Columns

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5-Oct","5-10")
df.head()
```

# Gender Destribution

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df,x = "Gender",palette={"male": "#ADD8E6", "female": "pink"})
ax.bar_label(ax.containers[0])
plt.title("Gender Destribution")
plt.show()
```

## From the above charts we have analysed that the number of females is more than the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb, annot = True)
plt.title("Relation between Parents Education and Childen's Score")
plt.show()
```

# From the above chart it was shown that the marks of the parents have a good empact on there kids education

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})

print(gb1)
```

```python
sns.heatmap(gb1 , annot = True)
plt.title("Relation between Parents Marital and Childen's Score")
plt.show()
```

# From the above chart it was shown that the marks of the parents have a negligible/no empact on there kids education

# Lunch type analysis

```python
gb3 = df.groupby("LunchType").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb3)
```

# Distribution of Ethinic Groups

```python
groupA = df.loc[(df['EthnicGroup']== "group A")].count()
groupB = df.loc[(df['EthnicGroup']== "group B")].count()
groupC = df.loc[(df['EthnicGroup']== "group C")].count()
groupD = df.loc[(df['EthnicGroup']== "group D")].count()
groupE = df.loc[(df['EthnicGroup']== "group E")].count()


mlist = ["groupA","groupB","groupC","groupD","groupE"]
l = [groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
plt.pie(l,labels = mlist,autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

# Gender vs All scores


```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
sns.boxplot(data=df, x='Gender', y='MathScore')
plt.title('Math Score')

plt.subplot(1, 3, 2)
sns.boxplot(data=df, x='Gender', y='ReadingScore')
plt.title('Reading Score')

plt.subplot(1, 3, 3)
sns.boxplot(data=df, x='Gender', y='WritingScore')
plt.title('Writing Score')

plt.tight_layout()
plt.show()
```

# Weekly Study Hours vs All Scores

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
sns.violinplot(x='WklyStudyHours', y='MathScore', data=df)
plt.title('Maths Score')


plt.subplot(1, 3, 2)
sns.violinplot(x='WklyStudyHours', y='ReadingScore', data=df)
plt.title('Reading Score')


plt.subplot(1, 3, 3)
sns.violinplot(x='WklyStudyHours', y='WritingScore', data=df)
plt.title('Writing Score')

plt.tight_layout()
plt.show()
```

# From the above charts, we can analyze the distribution of writing scores across different weekly study hour categories.
